## **IMPORTING LIBRARIES**

```python
# importing libraries.

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

# **IMPORTING THE DATASET**

```python
# Here we are loading netflix data file into the dataset using pandas.

dataset = pd.read_csv("/content/Dataset_master.xlsx - Netflix Movie Recommendation Ec (1).csv")

# Creating a list and appending the dataset in the form of string.

transactions = []

for i in range(0, len(dataset)):
  transactions.append([str(dataset.values[i, j]) for j in range(0, 12)])
```

# **APRIORI TRAINING ON DATASET**

```python
# Here simply, we are importing 'apriori' which is a algorithm for associtaion rule mining from 'apyori' which is a python library.

from apyori import apriori

# we are setting the values for the apriori.

rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

# **OUTPUT**

# **RAW RESULTS**

```python
# Now, we are converting into list and store it in the results.

results = list(rules)
```

```python
# Displying the raw results here

results
```

# **PROPER DISPLAY**

```python
# In the inspect funtion we are passing results in which it take the in the basis of suport values for the itemsets.

def inspect(results):
    movie_1         = [tuple(result[2][0][0])[0] for result in results]
    movie_2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(movie_1, movie_2, supports))

# Now, by using pandas we are creating/visualizing a table in which it shows the list of the frequent movies.

resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Movie 1', 'Movie 2', 'Support'])
```

```python
# Simply, displaying the all frequent itemsets in the table.

resultsinDataFrame
```

```python
#Finally, we are displaying the top most watched frequent movies from the dataset using APRIORI.

resultsinDataFrame.nlargest(n = 10,  columns = 'Support')
```
